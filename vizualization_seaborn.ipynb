```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns #agar seaborn ornatilagan bolsa pip install seaborn
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv",index_col = 0)
df.head()
```

```python
df.shape # bu datani ustun va qatorlar sonini chiqarib beradi
```

**Datamini seaborn yordamida chaqirib olish**

```python
sns.lineplot(data = df)
plt.show()
```

**Data ni matplotlib yordamida chizish**

```python
plt.plot(df)
plt.show()
```

**Grafigimizga style(stiller) beramiz**

```python
plt.figure(figsize=(16,6))
plt.title('Muzeylarga tashrif grafigi')
plt.xlabel('Sana')
plt.xticks(rotation=90) # x oqidagi yozuvlarni 90 gradusga  aylantiramiz.
plt.ylabel('Odamlar soni')
sns.lineplot(data = df)
plt.show()
```

Avila Adobe muzeyi uchun grafik chizamiz.

```python
plt.figure(figsize=(16,5))
plt.title("Avila Adobe muzeyiga tashrif buyuruvchilar")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.xticks(rotation=90)
sns.lineplot(data = df['Avila Adobe'])
plt.show()
```

**Bir nechtasini bitta grafikda tasvirlash uchun**

```python
plt.figure(figsize = (12,6))
plt.title('Muzeylarga tashrif buyuruvchilar soni ')
plt.xlabel('Sana')
plt.ylabel('Odamlar soni')
plt.xticks(rotation = 90)
sns.lineplot(data = df[['Avila Adobe','Chinese American Museum']])
plt.show()
```
